```python
import pandas as pd
import matplotlib.pyplot as plt  # Para personalizar el gráfico si es necesario

# Cargar el archivo Excel
file_path = '/Users/user/Downloads/BD CRM CAMBIO.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Verificar si la columna "CampaÒa" existe
columna_para_analizar = "CampaÒa"

if columna_para_analizar in df.columns:
    # Generar gráfico de barras con los 10 valores más comunes
    df[columna_para_analizar].value_counts().head(10).plot(kind='bar', color='skyblue', figsize=(10, 6))
    
    # Personalizar el gráfico
    plt.title(f"Top 10 valores de la columna '{columna_para_analizar}'", fontsize=14)
    plt.xlabel("Valores", fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje X
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Mostrar el gráfico
    plt.tight_layout()  # Ajustar el diseño para evitar etiquetas cortadas
    plt.show()
else:
    print(f"La columna '{columna_para_analizar}' no existe en el archivo.")
```
