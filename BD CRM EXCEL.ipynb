```python
import pandas as pd
import matplotlib.pyplot as plt  # Para personalizar el gráfico si es necesario

# Cargar el archivo Excel
file_path = '/Users/user/Downloads/BD CRM CAMBIO.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Verificar si la columna "CampaÒa" existe
columna_para_analizar = "CampaÒa"

if columna_para_analizar in df.columns:
    # Generar gráfico de barras con los 10 valores más comunes
    df[columna_para_analizar].value_counts().head(10).plot(kind='bar', color='skyblue', figsize=(10, 6))
    
    # Personalizar el gráfico
    plt.title(f"Top 10 valores de la columna '{columna_para_analizar}'", fontsize=14)
    plt.xlabel("Valores", fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje X
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Mostrar el gráfico
    plt.tight_layout()  # Ajustar el diseño para evitar etiquetas cortadas
    plt.show()
else:
    print(f"La columna '{columna_para_analizar}' no existe en el archivo.")
```

# Matplolib

```python
import pandas as pd
import matplotlib.pyplot as plt  # Para personalizar el gráfico si es necesario

# Cargar el archivo Excel
file_path = '/Users/conta/Desktop/BD Cambio Colombia.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Verificar si la columna "Plan de pago" existe
columna_para_analizar = "Plan de pago"

if columna_para_analizar in df.columns:
    # Generar gráfico de barras con los 10 valores más comunes
    df[columna_para_analizar].value_counts().head(3).plot(kind='bar', color='skyblue', figsize=(8, 4))
    
    # Personalizar el gráfico
    plt.title(f"Top 10 valores de la columna '{columna_para_analizar}'", fontsize=14)
    plt.xlabel("Valores", fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje X
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Mostrar el gráfico
    plt.tight_layout()  # Ajustar el diseño para evitar etiquetas cortadas
    plt.show()
else:
    print(f"La columna '{columna_para_analizar}' no existe en el archivo.")
```

Este gráfico muestra la cantidad de suscripciones que tiene plan de pago Corriente y plan de pago Recurrente

# Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el archivo Excel
file_path = '/Users/conta/Desktop/BD Cambio Colombia.xlsx'  # Ajusta la ruta si es necesario
df = pd.read_excel(file_path, engine='openpyxl')

# Verificar si las columnas necesarias existen
columna_x = "Tiempo suscripcion"
columna_y = "Valor"

if columna_x in df.columns and columna_y in df.columns:
    # Configurar estilo de seaborn
    sns.set(style="whitegrid")
    
    # Crear gráfico de dispersión
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        x=columna_x,
        y=columna_y,
        data=df,
        hue="Estado",  # Diferenciar los puntos según el estado
        palette="viridis",
        alpha=0.7
    )
    
    # Personalizar el gráfico
    plt.title(f"Relación entre '{columna_x}' y '{columna_y}'", fontsize=16)
    plt.xlabel(columna_x, fontsize=12)
    plt.ylabel(columna_y, fontsize=12)
    plt.legend(title="Estado", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    
    # Mostrar el gráfico
    plt.show()
else:
    print(f"Una o ambas columnas ('{columna_x}', '{columna_y}') no existen en el archivo.")
```

Con este grafico de dispersion puedo indentificar patrones entre la duración de la suscripción y el monto pagado ó detectar anomalias o datos inusuales

# Plotly

```python
import pandas as pd
import plotly.express as px

# Cargar el archivo Excel
file_path = '/Users/conta/Desktop/BD Cambio Colombia.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Verificar si las columnas necesarias existen
columna_x = "Tiempo suscripcion"  
columna_y = "Valor"  
columna_color = "Estado"  
if columna_x in df.columns and columna_y in df.columns and columna_color in df.columns:
    # Crear gráfico de dispersión interactivo
    fig = px.scatter(
        df,
        x=columna_x,
        y=columna_y,
        color=columna_color,  # Diferenciar por colores
        size="Valor",  
        hover_data=["Nombre", "Apellido", "Documento"],  # Para lo del mouse
        title="Relación entre Tiempo de Suscripción y Valor",
        labels={
            columna_x: "Tiempo de Suscripción (días)",
            columna_y: "Valor ($)",
        },
        template="plotly_white"
    )
    
    # Personalizar el diseño
    fig.update_traces(marker=dict(opacity=0.7, line=dict(width=1, color='DarkSlateGrey')))
    fig.update_layout(
        title_font_size=18,
        legend_title="Estado de Suscripción",
        xaxis_title="Tiempo de Suscripción (días)",
        yaxis_title="Valor ($)",
    )
    
    # Mostrar el gráfico
    fig.show()
else:
    print(f"Una o más columnas necesarias ('{columna_x}', '{columna_y}', '{columna_color}') no existen en el archivo.")
```
